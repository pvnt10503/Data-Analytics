# Outlier Treament
* Capping and Flooring
- impute all the values above 3*p99 and belowe 0.3*P1 -> replace outliers with 3*p99 for values above 3*p99 and 0.3*P1 for values below 0.3*p1
- Can use multipler instead of 3

```python
np.percentile(df.n_hot_rooms,[99])
```

```python
np.percentile(df.n_hot_rooms,[99])[0]
```

```python
uv = np.percentile(df.n_hot_rooms,[99])[0]
```

```python
df[(df.n_hot_rooms>uv)]
```

```python
df.n_hot_rooms[(df.n_hot_rooms>3*uv)]=3*uv 
```

```python
np.percentile(df.rainfall,[1])[0]
```

```python
lv = np.percentile(df.rainfall,[1])[0]
```

```python
df[(df.rainfall<lv)]
```

```python
df.rainfall[(df.rainfall<0.3*lv)]=0.3*lv
df.loc[df.rainfall<0.3*lv,'rainfall']=0.3*lv
df.head()
```
